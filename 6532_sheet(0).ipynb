```python
#Libraries used
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#0. Loading the iris dataset
iris_dataset = sns.load_dataset('iris')
```

```python
#1. Plot the histogram for each class.
plt.hist(iris_dataset['species'])
```

```python
#1(continued) Plot the histogram for each class.
figure = plt.figure(figsize=(15, 15))

pl0 = figure.add_subplot(2,2,1)
plt.hist(iris_dataset["sepal_length"], bins = 20)
plt.title("Sepal Length")

pl0 = figure.add_subplot(2,2,2)
plt.hist(iris_dataset["sepal_width"], bins = 20)
plt.title("Sepal Width")

pl0 = figure.add_subplot(2,2,3)
plt.hist(iris_dataset["petal_length"], bins = 20)
plt.title("Petal Length")

pl0 = figure.add_subplot(2,2,4)
plt.hist(iris_dataset["petal_width"], bins = 20)
plt.title("Petal Width")

plt.show()
```

```python
#2. Use scatter plot to plot every 2 attributes together.
figure1, pl = plt.subplots(4, 4, figsize=(18, 15))

sns.scatterplot(x = 'sepal_length', y = 'sepal_length', data=iris_dataset, hue='species', ax=pl[0,0])
sns.scatterplot(x = 'sepal_length', y = 'sepal_width', data=iris_dataset, hue='species', ax=pl[0][1])
sns.scatterplot(x = 'sepal_length', y = 'petal_length', data=iris_dataset, hue='species', ax=pl[0][2])
sns.scatterplot(x = 'sepal_length', y = 'petal_width', data=iris_dataset, hue='species', ax=pl[0][3])

sns.scatterplot(x = 'sepal_width', y = 'sepal_width', data=iris_dataset, hue='species', ax=pl[1][0])
sns.scatterplot(x = 'sepal_width', y = 'sepal_length', data=iris_dataset, hue='species', ax=pl[1][1])
sns.scatterplot(x = 'sepal_width', y = 'petal_length', data=iris_dataset, hue='species', ax=pl[1][2])
sns.scatterplot(x = 'sepal_width', y = 'petal_width', data=iris_dataset, hue='species', ax=pl[1][3])

sns.scatterplot(x = 'petal_length', y = 'petal_length', data=iris_dataset, hue='species', ax=pl[2][0])
sns.scatterplot(x = 'petal_length', y = 'sepal_length', data=iris_dataset, hue='species', ax=pl[2][1])
sns.scatterplot(x = 'petal_length', y = 'sepal_width', data=iris_dataset, hue='species', ax=pl[2][2])
sns.scatterplot(x = 'petal_length', y = 'petal_width', data=iris_dataset, hue='species', ax=pl[2][3])

sns.scatterplot(x = 'petal_width', y = 'petal_width', data=iris_dataset, hue='species', ax=pl[3][0])
sns.scatterplot(x = 'petal_width', y = 'sepal_length', data=iris_dataset, hue='species', ax=pl[3][1])
sns.scatterplot(x = 'petal_width', y = 'sepal_width', data=iris_dataset, hue='species', ax=pl[3][2])
sns.scatterplot(x = 'petal_width', y = 'petal_length', data=iris_dataset, hue='species', ax=pl[3][3])

plt.show()
```

```python
#3. Use 3D scatter plot to plot every 3 attributes together.
figure2 = plt.figure(figsize=(15,15))
pl2 = figure2.add_subplot(2,2,1, projection='3d')
x1 = iris_dataset['sepal_length']
y1 = iris_dataset['sepal_width']
z1 = iris_dataset['petal_length']
pl2.scatter(x1, y1, z1)
pl2.set_xlabel("sepal_length")
pl2.set_ylabel("sepal_width")
pl2.set_zlabel("petal_length")

pl2 = figure2.add_subplot(2,2,2, projection='3d')
x2 = iris_dataset['sepal_length']
y2 = iris_dataset['sepal_width']
z2 = iris_dataset['petal_width']
pl2.scatter(x2, y2, z2)
pl2.set_xlabel("sepal_length")
pl2.set_ylabel("sepal_width")
pl2.set_zlabel("petal_width")

pl2 = figure2.add_subplot(2,2,3, projection='3d')
x3 = iris_dataset['sepal_length']
y3 = iris_dataset['petal_width']
z3 = iris_dataset['petal_length']
pl2.scatter(x3, y3, z3)
pl2.set_xlabel("sepal_length")
pl2.set_ylabel("petal_width")
pl2.set_zlabel("petal_length")

pl2 = figure2.add_subplot(2,2,4, projection='3d')
x4 = iris_dataset['sepal_width']
y4= iris_dataset['petal_width']
z4 = iris_dataset['petal_length']
pl2.scatter(x4, y4, z4)
pl2.set_xlabel("sepal_width")
pl2.set_ylabel("petal_width")
pl2.set_zlabel("petal_length")

plt.show()
```

```python
#4. Show the correlation matrix between the attributes and comment on it.
iris_dataset.corr()

#Comments
#The diagonals' correlation is equal to 1 because each atttribute is perfectly correlated with itself.
#0.962865 between petal_length and petal_width means that they're strongly positively correlated.
#-0.117570 between sepal_lenth and sepal_width means that they're weakly negatively correlated.
```

```python
#4(continued) Plotting correlation heatmap.
sns.heatmap(iris_dataset.corr(),annot=True)
```

```python
#5. Use boxplots to show the quartiles and the median of the sepal length,
#sepal width, petal length and petal width for each kind of flowers.
sns.set()
sns.set_theme(style="whitegrid")
figure3, pl3 = plt.subplots(2, 2, figsize=(15,15))

sns.boxplot(x="species", y="sepal_length", data=iris_dataset, ax=pl3[0,0])
sns.boxplot(x="species", y="sepal_width", data=iris_dataset, ax=pl3[0,1])
sns.boxplot(x="species", y="petal_length", data=iris_dataset, ax=pl3[1,0])
sns.boxplot(x="species", y="petal_width", data=iris_dataset, ax=pl3[1,1])

plt.show()
```
